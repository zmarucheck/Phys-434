## Problem 0
#### Generation of Data to Play Around With

```python
import numpy as np
import scipy.stats as stats

t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
print(np.shape(bkgd))
```

## Problem 1
#### Part 1a) Make a scatter plot (not a line plot) of the first 1000 data points.

```python
import matplotlib.pyplot as plt

plt.scatter(np.linspace(0,1000,1000),bkgd[0:1000])
plt.title("Normal Data First 1000 Points")
plt.xlabel("Point Number")
plt.ylabel("Value")
plt.show()
```

#### Part 1b) Generalize your code so you can make a plot of any X contiguous points (e.g. 2000 points starting at point 500), and produce an example plot of a set of data somethere in the middle of your array.

```python
x = 2000
start = 500

plt.scatter(np.linspace(start,x+start,x),bkgd[start:x+start])
plt.title("Normal Data Points 500 through 2500")
plt.xlabel("Point Number")
plt.ylabel("Value")
plt.show()
```

#### Part 1c) Sometimes you want to sample the data, such as plotting every 100th point. Make a plot of the full data range, but only every 100th point.

```python
final_len = 1000
plt.scatter(np.linspace(0,100000,final_len),bkgd[np.arange(0,100000,100)])
plt.title("Normal Data Points Sampled for Every 100th Point")
plt.xlabel("Point Number")
plt.ylabel("Value")
plt.show()
```

## Problem 2
#### Part 2a) Make a 2d histogram plot. Read about options, and play with binsize so it is useful (too large a bin and everything washes out, too small and the sample noise dominates). Make sure you use a colorbar.

```python
import matplotlib.colors as colors

x = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
y = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
fig, ax = plt.subplots()
hh = ax.hist2d(x, y, bins=60, norm=colors.LogNorm())
fig.colorbar(hh[3], ax=ax)
plt.title("2D Histogram of 100k Points Normal vs Normal")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

#### Part 2b) Clearly explain what is being plotted in your plot (e.g. what does each color mean?)
The colors represent the density of the data found at that particular x and y segment, with darker blue colors representing areas with few to no points of data, lighter blue to green representing areas with middling amounts of data (\~10 to \~100 data points per 2D segment), and yellow representing the areas with the highest density of data (>\~120).

## Problem 3
#### Part 3a) Make an new time variable that folds at a period of  2  (the modulo operator is your friend, variably expressed as mod(), %, or rem()). Then make a scatter plot of all your data, but now folded. (Reminder, for more data you'd never actually want to plot all the points.)



```python
import math

plt.scatter(np.linspace(0,1000,1000),bkgd[0:1000]%(2*math.pi))
plt.title("Normal Data First 1000 Points Folded at a Period of 2pi")
plt.xlabel("Point Number")
plt.ylabel("Value")
plt.show()
```

```python
x = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
y = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
fig, ax = plt.subplots()
hh = ax.hist2d(x%(2*math.pi), y%(2*math.pi), bins=60, norm=colors.LogNorm())
fig.colorbar(hh[3], ax=ax)
plt.title("2D Histogram of 100k Points Normal vs Normal, Both Folded by 2pi")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
