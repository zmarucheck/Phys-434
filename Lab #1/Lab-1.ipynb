## Problem 2
#### A) The chi-squared distribution is comprised of the sum of 'k' different normally distributed random variables, each one independently squared, where 'k' is the degrees of freedom specified.

```python
# Part B
import matplotlib.pyplot as plt
from scipy.stats import chi2
import numpy as np

#plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(1, 1)

df = 55 #degrees of freedom, not sure what to set but 55 works as a default

x = np.linspace(chi2.ppf(0.01, df),
                chi2.ppf(0.99, df), 100)
ax.plot(x, chi2.pdf(x, df), label='chi^2')

r = chi2.rvs(df, size=100000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.5)
ax.legend(loc='best', frameon=False)
plt.show()
```

## Problem 3
#### A) Selecting 40 as value.
#### B) What is the chance that a value at or below 40 will be selected given the chi-squared distribution shown in Problem 2?
#### C) Integral from -infinity to 40 of f, where f is the chi-squared distribution defined above.

```python
# Part D
prob = chi2.cdf(40, 55)
print("Probability that a value at or below 40 will be selected is: " + str(prob))
# Part E
print("This is equivalent to a sigma of " + str(stats.norm.ppf(prob)))
```

## Problem 4
Probability crosses over the 0.5 mark and sigma changes from negative to positive as the value being measured passes 55 (which is the number of degrees of freedom and the seeming mean of the graph), although interestingly 55 exactly has a positive sigma and a probability greater than 0.5. As the value being measured increases above 55, the probability approaches 1 and sigma increases at an increasing pace. As the value being measured decreases below 55, the probability approaches 0 and the sigma decreases at an increasing pace.

```python
# Testing part of Problem 4, change n to value being checked and run to see the values of probability and sigma.
n = 150
probability = chi2.cdf(n, 55)
print("Probability: " + str(probability) + " and sigma: " + str(stats.norm.ppf(probability)))
```
