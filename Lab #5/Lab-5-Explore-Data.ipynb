# Correlation and covariance

```python
# Print covariance between a bunch of different things
labels = ["pt","eta","phi","mass","ee2","ee3","d2","angularity","t1","t2","t3","t21","t32","KtDeltaR"]
for i in labels:
    print("New first variable of: " + i)
    for j in labels:
        if (abs(np.cov(qcd[i],qcd[j])[0,1]) > 1 or abs(np.cov(higgs[i],higgs[j])[0,1]) > 1) and i != j:
            print("QCD covariance between " + i + " and " + j + " is: " + str(np.cov(qcd[i],qcd[j])[0,1]))
            print("Higgs covariance between " + i + " and " + j + " is: " + str(np.cov(higgs[i],higgs[j])[0,1]))
```

```python
# Print correlation between a bunch of different things
labels = ["pt","eta","phi","mass","ee2","ee3","d2","angularity","t1","t2","t3","t21","t32","KtDeltaR"]
for i in labels:
    print("New first variable of: " + i)
    for j in labels:
        if (abs(np.corrcoef(qcd[i],qcd[j])[0,1]) > 0.8 or abs(np.corrcoef(higgs[i],higgs[j])[0,1]) > 0.8) and i != j:
            print("QCD correlation between " + i + " and " + j + " is: " + str(np.corrcoef(qcd[i],qcd[j])[0,1]))
            print("Higgs correlation between " + i + " and " + j + " is: " + str(np.corrcoef(higgs[i],higgs[j])[0,1]))
```
